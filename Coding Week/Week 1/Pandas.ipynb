**Filtering rows of DF by column value**

```python
movies.head()
```

```python
movies.shape
```

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]
```

```python
is_long = movies.duration >= 200
```

```python
is_long.head()
```

```python
type(is_long)
```

```python
movies[movies.duration>=200]
```

**Applying Multiple Filter Criteria to a DF**

```python
movies[movies.duration>=200]
```

```python
movies[(movies.duration>=200) & (movies.genre == 'Drama')]
```

```python
movies[(movies.duration>=200) | (movies.genre == 'Drama')]
```

```python
 movies[movies.genre in ['Crime','Drama']]
```

```python
 movies[movies['genre'].isin(['Crime','Drama'])]
```

 **GroupBy in Pandas**

```python
drinks.head()
```

```python
drinks.beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks.groupby('continent')['beer_servings'].mean()
```

```python
# %matplotlib inline
# drinks.groupby('continent').mean().plot(kind='bar')
drinks.groupby('continent')['beer_servings'].mean().plot(kind='bar')
```
